# **Referencial Teórico**

<div align="justify"> <p> A análise de séries temporais constitui uma das abordagens mais relevantes para investigar fenômenos socioeconômicos que apresentam comportamento periódico ou evolutivo, como a dinâmica do emprego formal no Brasil. De acordo com Morettin e Toloi (2018), uma série temporal é formada por observações coletadas em intervalos regulares, nas quais cada valor mantém correlação com os anteriores, permitindo a identificação de padrões de tendência, sazonalidade, ciclos e ruído. Esse tipo de análise não apenas descreve eventos passados, mas também possibilita a elaboração de previsões e cenários prospectivos, o que se mostra essencial para a formulação de políticas públicas e estratégias organizacionais.

Entre as metodologias clássicas, destacam-se os modelos ARIMA (AutoRegressive Integrated Moving Average), popularizados por Box, Jenkins e Reinsel (2015), amplamente aplicados em contextos de séries estacionárias. A principal vantagem desses modelos é a flexibilidade para captar relações autorregressivas e médias móveis; contudo, sua limitação reside na necessidade de estacionariedade e na complexidade de parametrização, o que pode dificultar sua aplicação em séries mais curtas ou com elevada variabilidade.

Outra linha de abordagem é a dos métodos de suavização exponencial, como Holt-Winters, amplamente reconhecidos pela eficiência em capturar tendências e sazonalidades em dados econômicos (HYNDMAN; ATHANASOPOULOS, 2018). Tais métodos apresentam simplicidade operacional e bom desempenho em horizontes curtos, ainda que sejam sensíveis a choques abruptos. Em complemento, técnicas como médias móveis — simples, ponderadas ou exponenciais — oferecem recursos interpretativos valiosos para destacar tendências e reduzir ruídos, ainda que apresentem limitações na modelagem de relações complexas com variáveis externas.

As técnicas de decomposição clássica, por sua vez, permitem separar a série em seus componentes estruturais — tendência, sazonalidade e resíduo — favorecendo análises descritivas, comparações setoriais e diagnósticos regionais (CLEVELAND et al., 1990). Embora não sejam voltadas prioritariamente para previsão, constituem recurso importante para a caracterização do fenômeno em estudos exploratórios.

Mais recentemente, avanços no campo da Ciência de Dados têm possibilitado o desenvolvimento de modelos híbridos, que combinam métodos estatísticos tradicionais com técnicas de aprendizado profundo. Pesquisas publicadas no IEEE Xplore indicam que a integração de modelos como ARIMA com redes neurais recorrentes (LSTM e GRU) proporciona ganhos significativos de acurácia, sobretudo em séries não lineares ou com padrões de alta variabilidade (ZHANG, 2003; LI et al., 2021). Esses modelos apresentam como vantagens a flexibilidade e a capacidade de capturar dinâmicas complexas; entretanto, demandam maior volume de dados, infraestrutura computacional e maior complexidade interpretativa.

No caso específico do mercado de trabalho formal brasileiro, o Novo CAGED disponibiliza mensalmente informações sobre admissões, desligamentos, saldo e estoque de vínculos celetistas, constituindo fonte primária para análises de políticas de emprego (BRASIL, 2025). O Sumário Executivo funciona como régua de validação, apresentando sínteses nacionais, recortes setoriais, regionais e de perfil de trabalhadores (sexo, idade, escolaridade, faixas salariais). Embora esses relatórios assegurem confiabilidade e padronização, apresentam limitações de granularidade quando comparados aos microdados completos.

Diante disso, a solução proposta neste trabalho busca aliar a clareza interpretativa dos métodos estatísticos clássicos à capacidade descritiva da decomposição e da análise exploratória de dados. O pipeline proposto contempla tanto a caracterização dos saldos por setor, região e perfil quanto a proposição de indicadores sintéticos, de forma a gerar relatórios replicáveis e de fácil interpretação para gestores públicos e privados. Dessa forma, o projeto se ancora em uma tradição metodológica robusta e, ao mesmo tempo, mantém abertura para incorporar técnicas mais sofisticadas em fases posteriores, em consonância com práticas recentes discutidas na literatura internacional.





# **Diagrama de Solução**

### **1. Coleta de dados**

A etapa de coleta de dados será responsável por reunir e organizar as bases oficiais que servirão de insumo para todas as análises subsequentes. Serão utilizadas duas fontes principais disponibilizadas pelo Ministério do Trabalho e Emprego (MTE): o Sumário Executivo do Novo CAGED (PDF), que trará os resultados consolidados em nível nacional, setorial, regional e de perfil dos trabalhadores; e a planilha “3-tabelas_Junho de 2025” (Excel), que reunirá informações em maior nível de detalhamento, abrangendo séries históricas desde 2020, além de desagregações por unidades federativas, municípios e setores de atividade econômica.

A caracterização dessas bases permitirá compreender a natureza e a abrangência dos dados. O Sumário Executivo será utilizado como régua oficial de validação, fornecendo totais consolidados e indicadores de perfil (sexo, idade, escolaridade e faixa salarial). Já a planilha em Excel representará a base operacional do pipeline, contendo microdados agregados que possibilitarão comparações entre regiões, setores e faixas de tempo. Dessa forma, será possível conciliar consistência institucional com granularidade analítica.

Além da caracterização das fontes, a etapa de coleta será planejada de modo a garantir que os principais componentes de uma série temporal estejam preservados e possam ser identificados no processo analítico. A tendência será assegurada pela extensão da série histórica (2020–2025), permitindo observar movimentos de longo prazo, como a recuperação do mercado de trabalho após a pandemia. A sazonalidade será garantida pela periodicidade mensal contínua dos dados, essencial para identificar padrões recorrentes em determinados meses, como contratações no comércio no final do ano. Os ciclos econômicos, por sua vez, serão considerados a partir das séries setoriais e regionais, possibilitando a detecção de flutuações associadas a fases de expansão ou retração da economia. Por fim, o ruído será preservado por meio da coleta dos valores brutos, incluindo registros classificados como “não informados” ou oscilações abruptas, que posteriormente serão tratados nas etapas de pré-processamento e suavização.

Assim, a coleta de dados não terá apenas a função de reunir os insumos, mas também de estruturá-los de maneira a permitir que cada componente fundamental da série temporal seja identificado e mensurado. Essa organização inicial será decisiva para a qualidade das análises subsequentes, pois garantirá que tanto a consistência institucional quanto a riqueza analítica sejam preservadas desde o início do pipeline.


### **2. Processamento dos Dados (Pré-Processamento)**
Na etapa de pré-processamento, os dados coletados passarão por um conjunto estruturado de atividades que terão como objetivo garantir consistência, qualidade e adequação ao formato exigido pela análise de séries temporais. As principais atividades serão:

**Normalização de rótulos e cabeçalhos**: os nomes das variáveis e categorias serão padronizados por meio de técnicas de string normalization, como remoção de acentos, unificação de maiúsculas e minúsculas e substituição de espaços por underscores. Essa prática garantirá uniformidade e facilitará a manipulação computacional dos dados (MCKINNEY, 2017).

**Conversão e tipagem de variáveis**: as variáveis numéricas, como admissões, desligamentos, saldo e estoque, serão convertidas para tipos numéricos adequados (int64 ou float64), enquanto as variáveis de tempo serão transformadas em objetos datetime. Isso permitirá precisão nos cálculos e padronização da série temporal (MORETTIN; TOLOI, 2018).

**Construção da chave temporal**: será criada a variável periodo_ref, padronizada no formato “YYYY-MM”, que servirá como índice temporal fixo para toda a série. Essa chave assegurará que os dados sejam tratados em frequência mensal, condição essencial para análises de tendência e sazonalidade (SHUMWAY; STOFFER, 2017).

**Organização das dimensões geográficas e setoriais**: as informações de Brasil, regiões, unidades federativas e municípios serão harmonizadas conforme o padrão do IBGE, enquanto os setores de atividade serão alinhados à classificação CNAE. Essa etapa garantirá comparabilidade entre diferentes recortes e evitará duplicidades (PENG, 2020).

**Tratamento de valores faltantes**: valores ausentes serão tratados de acordo com a natureza da variável. Para séries agregadas, será aplicada a técnica de imputação por valores anteriores (forward fill), assegurando continuidade. Para variáveis categóricas, como escolaridade ou faixa salarial, será utilizada a categoria “não informado”, preservando a integridade do banco de dados (HYNDMAN; ATHANASOPOULOS, 2018).

**Checagem de consistência aritmética**: será verificado se o saldo corresponde à diferença entre admissões e desligamentos em todos os registros. Eventuais inconsistências serão corrigidas com base nas informações mais granulares disponíveis, reforçando a confiabilidade da base de dados (BRASIL, 2025).

**Criação de indicadores derivados**: sSerão gerados indicadores adicionais, como o saldo por 10 mil vínculos ativos, que permitirá comparações mais justas entre regiões de diferentes portes. Também serão construídos acumulados de 12 meses, prática recomendada para suavizar variações de curto prazo em séries econômicas (CLEVELAND et al., 1990).

**Separação de estoque com e sem ajuste**: as duas versões do estoque fornecidas pelo Novo CAGED — “com ajuste” e “sem ajuste” — serão mantidas, possibilitando análises comparativas e validação cruzada entre as metodologias de apuração. Essa prática aumentará a transparência e a robustez dos resultados.


### **3. Análise exploratória e visualização**
A análise exploratória (EDA) será conduzida com o objetivo de descrever padrões e revelar insights preliminares. Serão produzidas séries temporais agregadas para o saldo nacional (2020–2025), rankings de saldos por setor, UF e município, além de mapas coropléticos e heatmaps que evidenciarão a intensidade relativa da geração de empregos.

Serão também geradas visualizações para caracterizar o perfil dos vínculos (sexo, idade, escolaridade e faixa salarial), de acordo com os dados disponíveis no Sumário Executivo. Complementarmente, serão aplicadas técnicas de médias móveis e funções de autocorrelação (ACF e PACF), conforme recomendadas pela literatura, para identificar dependências temporais e padrões estruturais (SHUMWAY; STOFFER, 2017; HYNDMAN; ATHANASOPOULOS, 2018).


### **4. Modelagem**
A etapa de modelagem será conduzida com base em métodos estatísticos clássicos e de interpretação transparente, priorizando técnicas capazes de atender aos objetivos do projeto sem a necessidade de ferramentas preditivas excessivamente complexas. Inicialmente, serão aplicados os modelos Naive e Sazonal Naive, que projetarão os valores futuros de forma simples, utilizando como referência o último valor observado ou o valor do mesmo mês em anos anteriores. Apesar da simplicidade, esses modelos servirão como linha de base (benchmarks), permitindo avaliar a efetividade de abordagens mais elaboradas.

Em seguida, será realizada a decomposição clássica da série temporal, que permitirá separar os dados em três componentes principais: tendência, sazonalidade e resíduo. A análise desses elementos possibilitará identificar, por exemplo, se existe uma trajetória de crescimento sustentado do emprego formal após 2020, se determinados meses apresentam sazonalidade recorrente (como contratações no comércio em dezembro) e se picos isolados podem ser classificados como ruído.

Complementarmente, será aplicada a técnica de suavização exponencial de Holt-Winters, que incorporará parâmetros para capturar tendência e sazonalidade de forma adaptativa. Esse método permitirá elaborar previsões de curto prazo, como a estimativa de como os saldos poderão se comportar nos meses subsequentes a junho/2025, com base nos padrões históricos da série.

Por fim, será considerada a utilização de modelos ARIMA (AutoRegressive Integrated Moving Average), que combinam componentes autorregressivos, diferenciação e médias móveis. A aplicação dependerá da verificação da estacionariedade das séries, avaliada por meio de testes estatísticos específicos, e será realizada de forma exploratória em séries que apresentarem forte dependência temporal, como as do setor de serviços. Ainda que o foco do projeto não esteja na previsão de longo prazo, os modelos ARIMA contribuirão para consolidar análises de séries temporais curtas e médias.

Embora o escopo principal seja a utilização de métodos descritivos e de média complexidade, reconhece-se que abordagens híbridas, como a combinação de ARIMA com redes neurais recorrentes (LSTM e GRU), vêm sendo exploradas em estudos recentes indexados no IEEE Xplore. Contudo, tais técnicas não serão implementadas neste trabalho, sendo apenas consideradas como possibilidade futura em pesquisas posteriores.

### **5. Avaliação e Validação**

A etapa de avaliação e validação será conduzida em duas dimensões complementares. Na dimensão estatística, serão aplicadas métricas de erro como o Erro Absoluto Médio (MAE), a Raiz do Erro Quadrático Médio (RMSE) e o Erro Percentual Absoluto Médio (MAPE), que permitirão mensurar a proximidade entre os valores ajustados pelos modelos e os dados efetivamente observados. Adicionalmente, serão realizados testes de estacionariedade, como o Augmented Dickey-Fuller (ADF), que indicarão a adequação das séries ao uso de modelos ARIMA, além da análise da força da tendência e da sazonalidade, possibilitando concluir, por exemplo, se o setor de comércio apresenta sazonalidade mais forte que o setor industrial.

Já na dimensão institucional, a validação ocorrerá por meio da comparação dos resultados produzidos pelo pipeline com os totais oficiais divulgados no Sumário Executivo do Novo CAGED. Esse processo garantirá que os saldos, admissões e desligamentos processados coincidam com os números oficiais reportados pelo Ministério do Trabalho e Emprego. Dessa forma, será assegurada não apenas a consistência estatística dos modelos aplicados, mas também a conformidade das análises com a régua institucional de referência.

Essa abordagem integrada permitirá validar o pipeline em termos de qualidade metodológica e relevância prática, fornecendo análises reprodutíveis e úteis para gestores públicos e privados compreenderem a movimentação do emprego formal no Brasil em junho de 2025.

**3.2. Processamento dos Dados (Pré-Processamento)**

```python
import pandas as pd
import numpy as np
from pathlib import Path
import re, unicodedata

RAW_DIR = Path("/content/raw")
PROC_DIR = Path("/content/processed")
PROC_DIR.mkdir(exist_ok=True)
xlsx_path = next(RAW_DIR.glob("*.xlsx"))
print(f"📘 Usando arquivo: {xlsx_path.name}")


def normalize_text(s):
      if pd.isna(s):
        return ""
      s = str(s)
      s = "".join(ch for ch in unicodedata.normalize("NFKD", s) if not unicodedata.combining(ch))
      s = s.lower().strip().replace(" ", "_")
      s = re.sub(r"[^a-z0-9_]+", "", s)
      return s

def to_period_month(x):
    try:
        x = pd.to_datetime(x, errors="coerce")
        return x.dt.to_period("M").astype(str)
    except Exception:
        return pd.Series([np.nan]*len(x))

def num(x):
    return pd.to_numeric(x, errors="coerce")


targets = ["Tabela 1", "Tabela 2", "Tabela 4", "Tabela 6", "Tabela 6.1"]
xls = pd.ExcelFile(xlsx_path, engine="openpyxl")

dfs = {}
for t in targets:
    try:
        df = pd.read_excel(xls, sheet_name=t, header=None, engine="openpyxl")
        hdr = next(i for i,row in df.iterrows() if row.notna().sum()>=3)
        df = pd.read_excel(xls, sheet_name=t, header=hdr, engine="openpyxl")
        df.columns = [normalize_text(c) for c in df.columns]
        dfs[t] = df
        print(f" {t}: {df.shape[0]} linhas, {df.shape[1]} colunas | header={hdr}")
    except Exception as e:
        print(f" Falha ao ler {t}: {e}")


# 3.2.1 Normalização e tipagem
for name, df in dfs.items():
    df.columns = [normalize_text(c) for c in df.columns]
    for c in df.columns:
        if any(k in c for k in ["admis", "deslig", "saldo", "estoque"]):
            df[c] = num(df[c])
    dfs[name] = df


# 3.2.3 Criação da chave temporal periodo_ref
for name, df in dfs.items():
    if "periodo_ref" not in df.columns:
        # tenta deduzir do contexto (Junho de 2025)
        df["periodo_ref"] = "2025-06"
    else:
        df["periodo_ref"] = to_period_month(df["periodo_ref"])
    dfs[name] = df

# 3.2.4 Harmonização geográfica e setorial
def harmoniza_uf(v):
    if pd.isna(v): return np.nan
    v = str(v).strip().upper()
    mapa = {"BRASIL":"BR", "DF":"DISTRITO FEDERAL"}
    return mapa.get(v, v)

for name, df in dfs.items():
    for col in ["regiao","uf","unidade_da_federacao","municipio","estado"]:
        if col in df.columns:
            df[col] = df[col].map(harmoniza_uf)
    dfs[name] = df


# 3.2.5 Tratamento de valores faltantes
for name, df in dfs.items():
    df.ffill(inplace=True)
    dfs[name] = df

# 3.2.6 Checagem de consistência aritmética
for name, df in dfs.items():
    cols = df.columns
    if {"admissoes","desligamentos","saldo"}.issubset(cols):
        diff = (df["admissoes"] - df["desligamentos"]).fillna(0)
        erros = (abs(diff - df["saldo"]) > 1).sum()
        if erros:
            print(f"⚠️ {name}: {erros} linhas com inconsistência em saldo (corrigido).")
            df["saldo"] = diff
    dfs[name] = df

# 3.2.7 Indicadores derivados
for name, df in dfs.items():
    if {"saldo","estoque"}.issubset(df.columns):
        df["saldo_por_10mil"] = (df["saldo"] / df["estoque"]) * 10000
    dfs[name] = df


# 3.2.8 Separação de estoque com/sem ajuste
for name, df in dfs.items():
    estoque_cols = [c for c in df.columns if "estoque" in c]
    if len(estoque_cols) >= 2:
        base = df[[estoque_cols[0], estoque_cols[1]]].copy()
        base.columns = ["estoque_com_ajuste", "estoque_sem_ajuste"]
        dfs[name+"_estoque"] = base


# Exportação das bases tratadas
for name, df in dfs.items():
    out_path = PROC_DIR / f"{normalize_text(name)}_tratado.csv"
    df.to_csv(out_path, index=False, encoding="utf-8")
    print(f" {name} salvo em {out_path.name} ({df.shape[0]} linhas).")
```

```python
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from IPython.display import display, HTML
PROC_DIR = Path("/content/processed")

def _fmt_mil(x, pos):
    try:
        return f"{int(x):,}".replace(",", ".")
    except:
        return str(x)

def try_load(name_glob):
    files = list(PROC_DIR.glob(name_glob))
    return pd.read_csv(files[0]) if files else None

def first_text_col(df):
    if df is None:
        return None
    for c in df.columns:
        if df[c].dtype == "object" and not df[c].str.match(r"^\d+([.,]\d+)?$").all():
            return c
    return None

def cols_available(df, wanted):
    return [c for c in wanted if (df is not None and c in df.columns)]

def safe_num(df, cols):
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df


# Carregamento dos tratados
t1 = try_load("tabela_1_tratado.csv")
t2 = try_load("tabela_2_tratado.csv")
t3 = try_load("tabela_3_tratado.csv")
t4 = try_load("tabela_4_tratado.csv")
t6 = try_load("tabela_6_tratado.csv")
t61 = try_load("tabela_61_tratado.csv")

print("Arquivos carregados:")
for nm, df in [("tabela_1", t1), ("tabela_2", t2), ("tabela_3", t3), ("tabela_4", t4), ("tabela_6", t6), ("tabela_6_1", t61)]:
    print(f" • {nm}: {'ok' if df is not None else 'não encontrado'}")


# Relatório de Qualidade
def quality_report(df, name):
    if df is None:
        return None
    total = len(df)
    nulos = df.isna().sum().sort_values(ascending=False)
    tipos = df.dtypes
    # consistência saldo
    con_fields = cols_available(df, ["admissoes","desligamentos","saldo"])
    inconsist = None
    if set(con_fields) == {"admissoes","desligamentos","saldo"}:
        diff = (pd.to_numeric(df["admissoes"], errors="coerce") -
                pd.to_numeric(df["desligamentos"], errors="coerce"))
        inconsist = int(((diff - pd.to_numeric(df["saldo"], errors="coerce")).abs() > 1).sum())
    return {
        "name": name,
        "rows": total,
        "cols": df.shape[1],
        "nulls_top5": nulos.head(5).to_dict(),
        "has_periodo_ref": "periodo_ref" in df.columns,
        "consistencia_saldo_erros": inconsist
    }

qr = []
for nm, df in [("Tabela 1", t1), ("Tabela 2", t2), ("Tabela 3", t3), ("Tabela 4", t4), ("Tabela 6", t6), ("Tabela 6.1", t61)]:
    rep = quality_report(df, nm)
    if rep: qr.append(rep)

display(HTML("<h3>🔎 Relatório de Qualidade (amostra)</h3>"))
display(pd.DataFrame(qr))

# Validação com Sumário Executivo
sumario_ref = {"admissoes": 2139182, "desligamentos": 1972561, "saldo": 166621}  # jun/2025

def try_compare_totals(df, label="(desconhecido)"):
    if df is None:
        return None
    tmp = df.copy()
    num_cols = cols_available(tmp, ["admissoes","desligamentos","saldo"])
    tmp = safe_num(tmp, num_cols)
    total_guess = {k: (tmp[k].sum() if k in tmp.columns else np.nan) for k in num_cols}
    row = {"fonte": label}
    for k,v in total_guess.items():
        row[k] = int(v) if pd.notna(v) else np.nan
        row[k+"_ok_sumario"] = (int(v) == sumario_ref[k]) if pd.notna(v) and k in sumario_ref else None
    return row

checks = []
for (df, nm) in [(t1,"Tabela 1"), (t2,"Tabela 2")]:
    r = try_compare_totals(df, nm)
    if r: checks.append(r)

if checks:
    display(HTML("<h4>✅ Checagem contra Sumário Executivo (aproximação)</h4>"))
    display(pd.DataFrame(checks))
else:
    print(" Não foi possível comparar totais nessas tabelas — ok para avançar.")


```

```python
# ETAPA 3 — SÉRIES HISTÓRICAS (T6/T6.1): Médias móveis & STL


import re
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from statsmodels.tsa.seasonal import STL
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

PROC_DIR = Path("/content/processed")

def _fmt_mil(x, pos):
    try:
        return f"{int(x):,}".replace(",", ".")
    except:
        return str(x)


# 1) Carregar T6 / T6.1
t6_path  = PROC_DIR / "tabela_6_tratado.csv"
t61_path = PROC_DIR / "tabela_61_tratado.csv"
t6  = pd.read_csv(t6_path)  if t6_path.exists()  else None
t61 = pd.read_csv(t61_path) if t61_path.exists() else None

print("Disponibilidade das séries:")
print(" • Tabela 6 :", "ok" if t6 is not None else "não encontrada")
print(" • Tabela 6.1:", "ok" if t61 is not None else "não encontrada")

if t6 is None and t61 is None:
    raise FileNotFoundError("Não encontrei Tabela 6 nem 6.1 em /content/processed. Rode as etapas anteriores.")

df_wide = t61.copy() if t61 is not None else t6.copy()

# 2) Normalizações suaves
df_wide.ffill(inplace=True)


# 3) Detectar colunas de dimensão (ID) e colunas mensais (VALOR)

DIM_KEYS = [
    "regiao", "uf", "unidade_da_federacao", "estado", "municipio",
    "regiao_e_uf",
    "setor", "secao", "cnae", "grupamento_de_atividades", "grupamento_de_atividades_economicas", "grupamento"
]

def is_dim_col(col: str) -> bool:
    c = str(col).lower()
    return any(k in c for k in DIM_KEYS)

METRIC_NON_MONTH = ["admis", "deslig", "saldo", "estoque"]
MONTH_PATTERNS = [
    r"(janeiro|fevereiro|marco|abril|maio|junho|julho|agosto|setembro|outubro|novembro|dezembro)\s*20\d{2}",
    r"(janeiro|fevereiro|marco|abril|maio|junho|julho|agosto|setembro|outubro|novembro|dezembro)20\d{2}",
    r"20\d{2}[-_/]?((0[1-9])|(1[0-2]))",
    r"20\d{2}((0[1-9])|(1[0-2]))",
    r"\b(2020|2021|2022|2023|2024|2025)[-_]?(01|02|03|04|05|06|07|08|09|10|11|12)\b"
]

def looks_like_month_col(c: str) -> bool:
    cl = str(c).lower()
    if any(k in cl for k in ["unnamed"]):
        pass
    for pat in MONTH_PATTERNS:
        if re.search(pat, cl):
            return True
    return False


dim_cols = [c for c in df_wide.columns if is_dim_col(str(c))]
month_cols = [c for c in df_wide.columns if looks_like_month_col(str(c))]

def classify_month_metric(col: str) -> str:
    cl = str(col).lower()
    if "saldo" in cl:
        return "saldo"
    if "admis" in cl:
        return "admissoes"
    if "deslig" in cl:
        return "desligamentos"
    if "estoque" in cl:
        return "estoque"
    return "desconhecida"

month_metric_map = {c: classify_month_metric(c) for c in month_cols}

if not month_cols:
    raise AssertionError("Não identifiquei colunas mensais em T6/6.1. Verifique o cabeçalho e a etapa de tratamento.")

# 4) Escolher qual métrica usar (prioriza SALDO)
saldo_like_cols = [c for c in month_cols if month_metric_map[c] == "saldo"]
ad_cols        = [c for c in month_cols if month_metric_map[c] == "admissoes"]
des_cols       = [c for c in month_cols if month_metric_map[c] == "desligamentos"]

if saldo_like_cols:
    chosen_cols = saldo_like_cols
    chosen_metric = "saldo"
    print(f" Usando colunas mensais detectadas como SALDO ({len(chosen_cols)} colunas).")
elif ad_cols and des_cols:
    chosen_metric = "saldo_calc"
    chosen_cols = sorted(set(ad_cols + des_cols), key=lambda x: str(x))
    print(f" Não encontrei 'saldo' explícito. Vou calcular saldo = admissões - desligamentos.")
else:
    unk = [c for c in month_cols if month_metric_map[c] == "desconhecida"]
    if unk:
        chosen_cols = unk
        chosen_metric = "valor_mensal"
        print(f" Usando colunas mensais desconhecidas como valor: {len(unk)} colunas (fallback).")
    else:
        raise AssertionError("Não consegui determinar colunas mensais de saldo nem parear admissões/desligamentos.")


# 5) Funções de parsing do período (YYYY-MM)
MONTH_PT = {
    "janeiro":"01","fevereiro":"02","marco":"03","março":"03","abril":"04","maio":"05",
    "junho":"06","julho":"07","agosto":"08","setembro":"09","outubro":"10","novembro":"11","dezembro":"12"
}

def parse_period_from_col(colname: str) -> str | None:
    s = str(colname).lower().strip()


    m = re.search(r"(janeiro|fevereiro|marco|março|abril|maio|junho|julho|agosto|setembro|outubro|novembro|dezembro)\s*[-_/]*\s*(20\d{2})", s)
    if m:
        mm = MONTH_PT[m.group(1)]
        yyyy = m.group(2)
        return f"{yyyy}-{mm}"

    m2 = re.search(r"(20\d{2})[-_/]?(0[1-9]|1[0-2])", s)
    if m2:
        yyyy, mm = m2.group(1), m2.group(2)
        return f"{yyyy}-{mm}"

    m3 = re.search(r"(20\d{2})\s+(0[1-9]|1[0-2])", s)
    if m3:
        yyyy, mm = m3.group(1), m3.group(2)
        return f"{yyyy}-{mm}"

    return None

col_period = {c: parse_period_from_col(c) for c in chosen_cols}
chosen_cols = [c for c in chosen_cols if col_period.get(c) is not None]
assert chosen_cols, "As colunas selecionadas não geraram períodos válidos."


# 6) Se necessário, calcular saldo = admissões - desligamentos
def build_long_from_ad_des(df_wide):
    ad_map = {}
    des_map = {}
    for c in ad_cols:
        p = parse_period_from_col(c)
        if p: ad_map.setdefault(p, []).append(c)
    for c in des_cols:
        p = parse_period_from_col(c)
        if p: des_map.setdefault(p, []).append(c)

    dim_keep = [c for c in df_wide.columns if c in dim_cols]
    base = df_wide[dim_keep].copy()
    frames = []
    for p in sorted(set(ad_map.keys()).intersection(des_map.keys())):
        ad_sum  = df_wide[ad_map[p]].apply(pd.to_numeric, errors="coerce").sum(axis=1)
        des_sum = df_wide[des_map[p]].apply(pd.to_numeric, errors="coerce").sum(axis=1)
        saldo_p = ad_sum - des_sum
        tmp = base.copy()
        tmp["periodo_ref"] = p
        tmp["valor"] = saldo_p
        frames.append(tmp)
    if not frames:
        raise AssertionError("Não foi possível parear admissões e desligamentos por período.")
    return pd.concat(frames, ignore_index=True)


# 7) Montar dataframe LONG (periodo_ref, dims..., valor)
dim_keep = [c for c in df_wide.columns if c in dim_cols]
if chosen_metric == "saldo_calc":
    long_df = build_long_from_ad_des(df_wide)
else:
    sub = df_wide[dim_keep + chosen_cols].copy()
    for c in chosen_cols:
        sub[c] = pd.to_numeric(sub[c], errors="coerce")
    m = sub.melt(id_vars=dim_keep, value_vars=chosen_cols, var_name="col_mes", value_name="valor")
    m["periodo_ref"] = m["col_mes"].map(col_period)
    long_df = m.drop(columns=["col_mes"]).dropna(subset=["periodo_ref"])

long_df["periodo_ref"] = pd.PeriodIndex(long_df["periodo_ref"], freq="M").astype(str)


# 8) Cria chaves e indicadores derivados
def pick_key_cols(df):
    prefs = [
        ["uf", "setor"], ["uf", "secao"], ["uf", "grupamento_de_atividades_economicas"], ["uf"],
        ["regiao", "setor"], ["regiao", "secao"], ["regiao", "grupamento_de_atividades_economicas"], ["regiao"],
        ["municipio", "setor"], ["municipio"]
    ]
    for p in prefs:
        if all(c in df.columns for c in p):
            return p
    text_cols = [c for c in df.columns if c in dim_cols]
    return text_cols[:1] if text_cols else []

key_cols = pick_key_cols(long_df)
if key_cols:
    long_df["serie_key"] = long_df[key_cols].astype(str).agg(" | ".join, axis=1)
else:
    long_df["serie_key"] = "TOTAL"

agg_df = (long_df
          .groupby(["serie_key", "periodo_ref"], as_index=False)["valor"].sum()
          .sort_values(["serie_key", "periodo_ref"]))

def add_rolling(df):
    df = df.sort_values("periodo_ref")
    df["mm3"]  = df["valor"].rolling(3, min_periods=1).mean()
    df["mm6"]  = df["valor"].rolling(6, min_periods=1).mean()
    df["mm12"] = df["valor"].rolling(12, min_periods=1).mean()
    return df

agg_df = (agg_df
          .groupby("serie_key", group_keys=False)
          .apply(add_rolling)
          .reset_index(drop=True))

out_long = PROC_DIR / "series_long_t6_t61.csv"
agg_df.to_csv(out_long, index=False, encoding="utf-8")
print(f" Séries (long) salvas em: {out_long}")


# 9) Funções de plot
def plot_series(df, key=None, cols=("valor","mm3","mm6","mm12"), title_prefix="Série"):
    d = df if key is None else df[df["serie_key"] == key]
    if d.empty:
        print("Nada a plotar.")
        return
    x = pd.PeriodIndex(d["periodo_ref"], freq="M").to_timestamp()
    for c in cols:
        if c in d.columns:
            plt.figure(figsize=(9,4.5))
            plt.plot(x, d[c])
            plt.gca().yaxis.set_major_formatter(FuncFormatter(_fmt_mil))
            plt.title(f"{title_prefix} — {key if key else 'TOTAL (todas as chaves)'} — {c}")
            plt.xlabel("Período"); plt.ylabel("Postos")
            plt.tight_layout(); plt.show()

def plot_stl(df, key=None):
    d = df if key is None else df[df["serie_key"] == key]
    if d.empty:
        print("Nada para STL.")
        return
    d = d.copy().sort_values("periodo_ref")
    idx = pd.PeriodIndex(d["periodo_ref"], freq="M").to_timestamp()
    y = pd.Series(d["valor"].values, index=idx).asfreq("MS")
    y = y.interpolate()
    stl = STL(y, period=12, robust=True).fit()
    for comp_name, comp in [("Observado", stl.observed), ("Tendência", stl.trend),
                            ("Sazonal", stl.seasonal), ("Resíduo", stl.resid)]:
        plt.figure(figsize=(9,4))
        plt.plot(comp.index, comp.values)
        plt.gca().yaxis.set_major_formatter(FuncFormatter(_fmt_mil))
        plt.title(f"STL — {comp_name} — {key if key else 'TOTAL'}")
        plt.xlabel("Período"); plt.ylabel("Valor")
        plt.tight_layout(); plt.show()

def plot_acf_pacf(df, key=None, lags=24):
    d = df if key is None else df[df["serie_key"] == key]
    if d.empty:
        print("Nada para ACF/PACF.")
        return
    d = d.copy().sort_values("periodo_ref")
    idx = pd.PeriodIndex(d["periodo_ref"], freq="M").to_timestamp()
    y = pd.Series(d["valor"].values, index=idx).asfreq("MS").fillna(0)

    # ACF
    plt.figure(figsize=(9,4))
    plot_acf(y, lags=min(lags, len(y)-1))
    plt.title(f"ACF — {key if key else 'TOTAL'}")
    plt.tight_layout(); plt.show()

    # PACF
    plt.figure(figsize=(9,4))
    plot_pacf(y, lags=min(lags, len(y)-1), method="ywm")
    plt.title(f"PACF — {key if key else 'TOTAL'}")
    plt.tight_layout(); plt.show()


# 10) Exemplos de uso rápido
# a) Escolhe uma série-chave para demonstrar (TOTAL agregado ou a primeira chave)
sample_key = agg_df["serie_key"].iloc[0] if not agg_df.empty else None
print("Exemplo de chave para plot:", sample_key)

# b) Plot da série (valor e MMs)
plot_series(agg_df, key=sample_key, cols=("valor","mm3","mm6","mm12"), title_prefix="Emprego Formal (Saldo)")

# c) Decomposição STL
plot_stl(agg_df, key=sample_key)

# d) ACF / PACF
plot_acf_pacf(agg_df, key=sample_key, lags=24)
```

```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

REGION_NAMES = {
    "norte","nordeste","sudeste","sul","centro-oeste","centro_oeste","centro oeste"
}

def _fmt_mil(x, pos):
    try:
        return f"{int(x):,}".replace(",", ".")
    except:
        return str(x)

def first_text_col(df):
    if df is None or df.empty:
        return None
    for c in df.columns:
        if df[c].dtype == "object":
            # ignora colunas que parecem somente números
            if not df[c].astype(str).str.fullmatch(r"\s*[\d\.,]+\s*").all():
                return c
    return df.columns[0]

def safe_num(df, cols=None):
    if df is None: return df
    cols = cols or df.columns
    for c in cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")
    return df

def pick_label_col(df, candidates_keywords):
    cands = [c for c in df.columns if any(k in c for k in candidates_keywords)]
    return cands[0] if cands else first_text_col(df)

def pick_saldo_col(df):
    cols = list(df.columns)

    # 1) saldo exato
    if "saldo" in cols:
        return "saldo", "saldo (coluna exata)"

    # 2) qualquer coluna contendo 'saldo'
    saldo_like = [c for c in cols if "saldo" in c]
    if saldo_like:
        return saldo_like[0], f"saldo detectado: {saldo_like[0]}"

    # 3) calcular a partir de admissoes - desligamentos
    if "admissoes" in cols and "desligamentos" in cols:
        df["_saldo_calc"] = pd.to_numeric(df["admissoes"], errors="coerce") - pd.to_numeric(df["desligamentos"], errors="coerce")
        return "_saldo_calc", "saldo calculado (admissoes - desligamentos)"

    # 4) fallback: colunas mensais de junho/2025
    month_patterns = [r"junho\s*2025", r"junho2025", r"2025[-_/]?06"]
    month_cols = [c for c in cols if any(re.search(p, c) for p in month_patterns)]
    # evitar usar admissoes/desligamentos como saldo
    month_cols = [c for c in month_cols if not any(k in c for k in ["admis", "deslig"])]
    if month_cols:
        return month_cols[0], f"saldo (fallback mensal: {month_cols[0]})"

    # 5) 1ª coluna numérica “promissora”
    num_candidates = []
    for c in cols:
        ser = pd.to_numeric(df[c], errors="coerce")
        if ser.notna().sum() >= 3:
            num_candidates.append((c, ser.abs().sum()))
    if num_candidates:
        num_candidates.sort(key=lambda x: -x[1])
        return num_candidates[0][0], f"saldo (fallback genérico: {num_candidates[0][0]})"

    return None, "não foi possível identificar uma coluna de saldo"


# TABELA 1 — Saldo por Grupamento de Atividades
if t1 is not None:
    # Find the correct column name for activity group from processed data
    activity_col = None
    for col in t1.columns:
        if "grupamento" in col or "atividades" in col or "setor" in col:
            activity_col = col
            break
    if activity_col is None:
         activity_col = first_text_col(t1)

    # Use the processed dataframe directly and select/rename columns
    plot_df = t1[[activity_col, "unnamed_4"]].copy() # Use 'unnamed_4' for saldo
    plot_df = plot_df.rename(columns={"unnamed_4": "saldo"})

    # Convert saldo to numeric and drop NaNs
    plot_df["saldo"] = pd.to_numeric(plot_df["saldo"], errors="coerce")
    plot_df = plot_df.dropna(subset=["saldo"])

    # Filter out rows where activity group is empty or contains non-activity text
    plot_df = plot_df[plot_df[activity_col].astype(str).str.len()>0]
    plot_df = plot_df[~plot_df[activity_col].astype(str).str.contains("fonte|varia", case=False)]


    plot_df = plot_df.sort_values("saldo", ascending=True)

    plt.figure(figsize=(9,6))
    plt.barh(plot_df[activity_col], plot_df["saldo"])
    plt.gca().xaxis.set_major_formatter(FuncFormatter(_fmt_mil))
    plt.title("Saldo por Grupamento de Atividades — Jun/2025")
    plt.xlabel("Saldo (postos)")
    plt.ylabel("Grupamento")
    plt.tight_layout()
    plt.show()


# TABELA 2 —  Região × Grupamento (T2)
if t2 is not None:

    reg_col = "regiao" if "regiao" in t2.columns else None
    grp_col = pick_label_col(t2, ["grupamento","atividades","secao","setor","cnae"]) if any(k in " ".join(t2.columns) for k in ["grupamento","atividades","secao","setor","cnae"]) else None
    uf_col  = "uf" if "uf" in t2.columns else None
    reu_col = "regiao_e_uf" if "regiao_e_uf" in t2.columns else None
    val_col, how2 = pick_saldo_col(t2)
    if val_col is None:
        raise KeyError("Tabela 2: não foi possível identificar uma coluna de valores (saldo/menal).")

    t2 = safe_num(t2.copy(), [val_col])

    # Caso A: heatmap Região x Grupamento
    if reg_col and grp_col and reg_col in t2.columns and grp_col in t2.columns:
        pv = t2.pivot_table(index=reg_col, columns=grp_col, values=val_col, aggfunc="sum")
        plt.figure(figsize=(10,5))
        plt.imshow(pv.fillna(0).values, aspect="auto")
        plt.xticks(range(pv.shape[1]), pv.columns, rotation=45, ha="right")
        plt.yticks(range(pv.shape[0]), pv.index)
        plt.title("Saldo por Região × Grupamento — Jun/2025 (auto)")
        plt.colorbar()
        plt.tight_layout()
        plt.show()
        print("T2 -> Heatmap (Região × Grupamento).")

    # Caso B: coluna 'regiao_e_uf'
    elif reu_col and reu_col in t2.columns:
        lab = reu_col
        tmp = t2[[lab, val_col]].dropna().copy()
        tmp[lab] = tmp[lab].astype(str).str.strip().str.lower()

        # separa Regiões
        tmp["eh_regiao"] = tmp[lab].isin(REGION_NAMES)
        has_regions = tmp["eh_regiao"].any()

        if has_regions:
            por_regiao = tmp[tmp["eh_regiao"]].groupby(lab, as_index=False)[val_col].sum()
            por_regiao = por_regiao.sort_values(val_col, ascending=True)

            plt.figure(figsize=(8,5))
            plt.barh(por_regiao[lab].str.title(), por_regiao[val_col])
            plt.gca().xaxis.set_major_formatter(FuncFormatter(_fmt_mil))
            plt.title("Saldo por Região — (derivado de 'regiao_e_uf')")
            plt.xlabel("Saldo"); plt.ylabel("Região")
            plt.tight_layout(); plt.show()


        # Top UFs (filtra o que não é região)
        por_uf = tmp[~tmp["eh_regiao"]].groupby(lab, as_index=False)[val_col].sum()
        if not por_uf.empty:
            por_uf = por_uf.sort_values(val_col, ascending=True).tail(15)
            plt.figure(figsize=(8,6))
            plt.barh(por_uf[lab].str.upper(), por_uf[val_col])
            plt.gca().xaxis.set_major_formatter(FuncFormatter(_fmt_mil))
            plt.title("Top 15 UFs por Saldo")
            plt.xlabel("Saldo"); plt.ylabel("UF")
            plt.tight_layout(); plt.show()


    # Caso C: apenas regiao + valor
    elif reg_col and reg_col in t2.columns:
        tmp = t2[[reg_col, val_col]].dropna().copy()
        tmp = tmp.groupby(reg_col, as_index=False)[val_col].sum().sort_values(val_col, ascending=True)
        plt.figure(figsize=(8,5))
        plt.barh(tmp[reg_col], tmp[val_col])
        plt.gca().xaxis.set_major_formatter(FuncFormatter(_fmt_mil))
        plt.title("Saldo por Região — Jun/2025")
        plt.xlabel("Saldo"); plt.ylabel("Região")
        plt.tight_layout(); plt.show()


    # Caso D: apenas uf + valor
    elif uf_col and uf_col in t2.columns:
        tmp = t2[[uf_col, val_col]].dropna().copy()
        tmp = tmp.groupby(uf_col, as_index=False)[val_col].sum().sort_values(val_col, ascending=True).tail(15)
        plt.figure(figsize=(8,6))
        plt.barh(tmp[uf_col], tmp[val_col])
        plt.gca().xaxis.set_major_formatter(FuncFormatter(_fmt_mil))
        plt.title("Top 15 UFs por Saldo — Jun/2025")
        plt.xlabel("Saldo"); plt.ylabel("UF")
        plt.tight_layout(); plt.show()

    # Caso E: fallback genérico
    else:
        lab = first_text_col(t2)
        if lab is None:
            print("T2 -> Não há coluna de rótulo para plotar. Gráfico ignorado.")
        else:
            tmp = t2[[lab, val_col]].dropna().copy()
            tmp = tmp.groupby(lab, as_index=False)[val_col].sum()
            tmp = tmp.sort_values(val_col, ascending=True).tail(20)
            plt.figure(figsize=(9,6))
            plt.barh(tmp[lab].astype(str), tmp[val_col])
            plt.gca().xaxis.set_major_formatter(FuncFormatter(_fmt_mil))
            plt.title(f"Ranking — {lab} × valor (fallback T2)")
            plt.xlabel("Valor"); plt.ylabel(lab)
            plt.tight_layout(); plt.show()
```

```python
# ETAPA 3 — SÉRIES HISTÓRICAS: Médias móveis & STL


import re
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
from statsmodels.tsa.seasonal import STL
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

PROC_DIR = Path("/content/processed")

def _fmt_mil(x, pos):
    try:
        return f"{int(x):,}".replace(",", ".")
    except:
        return str(x)


t6_path  = PROC_DIR / "tabela_6_tratado.csv"
t61_path = PROC_DIR / "tabela_61_tratado.csv"
t6  = pd.read_csv(t6_path)  if t6_path.exists()  else None
t61 = pd.read_csv(t61_path) if t61_path.exists() else None

print("Disponibilidade das séries:")
print(" • Tabela 6 :", "ok" if t6 is not None else "não encontrada")
print(" • Tabela 6.1:", "ok" if t61 is not None else "não encontrada")

if t6 is None and t61 is None:
    raise FileNotFoundError("Não encontrei Tabela 6 nem 6.1 em /content/processed. Rode as etapas anteriores.")

df_wide = t61.copy() if t61 is not None else t6.copy()


df_wide.ffill(inplace=True)
DIM_KEYS = [
    "regiao", "uf", "unidade_da_federacao", "estado", "municipio",
    "regiao_e_uf",
    "setor", "secao", "cnae", "grupamento_de_atividades", "grupamento_de_atividades_economicas", "grupamento"
]

def is_dim_col(col: str) -> bool:
    c = str(col).lower()
    return any(k in c for k in DIM_KEYS)

METRIC_NON_MONTH = ["admis", "deslig", "saldo", "estoque"]
MONTH_PATTERNS = [
    r"(janeiro|fevereiro|marco|abril|maio|junho|julho|agosto|setembro|outubro|novembro|dezembro)\s*20\d{2}",
    r"(janeiro|fevereiro|marco|abril|maio|junho|julho|agosto|setembro|outubro|novembro|dezembro)20\d{2}",
    r"20\d{2}[-_/]?((0[1-9])|(1[0-2]))",
    r"20\d{2}((0[1-9])|(1[0-2]))",
    r"\b(2020|2021|2022|2023|2024|2025)[-_]?(01|02|03|04|05|06|07|08|09|10|11|12)\b"
]

def looks_like_month_col(c: str) -> bool:
    cl = str(c).lower()

    if any(k in cl for k in ["unnamed"]):
        pass
    for pat in MONTH_PATTERNS:
        if re.search(pat, cl):
            return True
    return False

dim_cols = [c for c in df_wide.columns if is_dim_col(str(c))]
month_cols = [c for c in df_wide.columns if looks_like_month_col(str(c))]

def classify_month_metric(col: str) -> str:
    cl = str(col).lower()
    if "saldo" in cl:
        return "saldo"
    if "admis" in cl:
        return "admissoes"
    if "deslig" in cl:
        return "desligamentos"
    if "estoque" in cl:
        return "estoque"
    return "desconhecida"

month_metric_map = {c: classify_month_metric(c) for c in month_cols}

if not month_cols:
    raise AssertionError("Não identifiquei colunas mensais em T6/6.1. Verifique o cabeçalho e a etapa de tratamento.")
saldo_like_cols = [c for c in month_cols if month_metric_map[c] == "saldo"]
ad_cols        = [c for c in month_cols if month_metric_map[c] == "admissoes"]
des_cols       = [c for c in month_cols if month_metric_map[c] == "desligamentos"]

if saldo_like_cols:
    chosen_cols = saldo_like_cols
    chosen_metric = "saldo"
    print(f" Usando colunas mensais detectadas como SALDO ({len(chosen_cols)} colunas).")
elif ad_cols and des_cols:
    chosen_metric = "saldo_calc"
    chosen_cols = sorted(set(ad_cols + des_cols), key=lambda x: str(x))
    print(f" Não encontrei 'saldo' explícito. Vou calcular saldo = admissões - desligamentos.")
else:
    unk = [c for c in month_cols if month_metric_map[c] == "desconhecida"]
    if unk:
        chosen_cols = unk
        chosen_metric = "valor_mensal"
        print(f" Usando colunas mensais desconhecidas como valor: {len(unk)} colunas (fallback).")
    else:
        raise AssertionError("Não consegui determinar colunas mensais de saldo nem parear admissões/desligamentos.")

MONTH_PT = {
    "janeiro":"01","fevereiro":"02","marco":"03","março":"03","abril":"04","maio":"05",
    "junho":"06","julho":"07","agosto":"08","setembro":"09","outubro":"10","novembro":"11","dezembro":"12"
}

def parse_period_from_col(colname: str) -> str | None:
    s = str(colname).lower().strip()

    m = re.search(r"(janeiro|fevereiro|marco|março|abril|maio|junho|julho|agosto|setembro|outubro|novembro|dezembro)\s*[-_/]*\s*(20\d{2})", s)
    if m:
        mm = MONTH_PT[m.group(1)]
        yyyy = m.group(2)
        return f"{yyyy}-{mm}"

    m2 = re.search(r"(20\d{2})[-_/]?(0[1-9]|1[0-2])", s)
    if m2:
        yyyy, mm = m2.group(1), m2.group(2)
        return f"{yyyy}-{mm}"

    m3 = re.search(r"(20\d{2})\s+(0[1-9]|1[0-2])", s)
    if m3:
        yyyy, mm = m3.group(1), m3.group(2)
        return f"{yyyy}-{mm}"

    return None

col_period = {c: parse_period_from_col(c) for c in chosen_cols}
chosen_cols = [c for c in chosen_cols if col_period.get(c) is not None]
assert chosen_cols, "As colunas selecionadas não geraram períodos válidos."

def build_long_from_ad_des(df_wide):
    ad_map = {}
    des_map = {}
    for c in ad_cols:
        p = parse_period_from_col(c)
        if p: ad_map.setdefault(p, []).append(c)
    for c in des_cols:
        p = parse_period_from_col(c)
        if p: des_map.setdefault(p, []).append(c)

    dim_keep = [c for c in df_wide.columns if c in dim_cols]
    base = df_wide[dim_keep].copy()
    frames = []
    for p in sorted(set(ad_map.keys()).intersection(des_map.keys())):
        ad_sum  = df_wide[ad_map[p]].apply(pd.to_numeric, errors="coerce").sum(axis=1)
        des_sum = df_wide[des_map[p]].apply(pd.to_numeric, errors="coerce").sum(axis=1)
        saldo_p = ad_sum - des_sum
        tmp = base.copy()
        tmp["periodo_ref"] = p
        tmp["valor"] = saldo_p
        frames.append(tmp)
    if not frames:
        raise AssertionError("Não foi possível parear admissões e desligamentos por período.")
    return pd.concat(frames, ignore_index=True)


dim_keep = [c for c in df_wide.columns if c in dim_cols]
if chosen_metric == "saldo_calc":
    long_df = build_long_from_ad_des(df_wide)
else:
    sub = df_wide[dim_keep + chosen_cols].copy()
    for c in chosen_cols:
        sub[c] = pd.to_numeric(sub[c], errors="coerce")
    m = sub.melt(id_vars=dim_keep, value_vars=chosen_cols, var_name="col_mes", value_name="valor")
    m["periodo_ref"] = m["col_mes"].map(col_period)
    long_df = m.drop(columns=["col_mes"]).dropna(subset=["periodo_ref"])


long_df["periodo_ref"] = pd.PeriodIndex(long_df["periodo_ref"], freq="M").astype(str)
def pick_key_cols(df):
    prefs = [
        ["uf", "setor"], ["uf", "secao"], ["uf", "grupamento_de_atividades_economicas"], ["uf"],
        ["regiao", "setor"], ["regiao", "secao"], ["regiao", "grupamento_de_atividades_economicas"], ["regiao"],
        ["municipio", "setor"], ["municipio"]
    ]
    for p in prefs:
        if all(c in df.columns for c in p):
            return p
    text_cols = [c for c in df.columns if c in dim_cols]
    return text_cols[:1] if text_cols else []

key_cols = pick_key_cols(long_df)
if key_cols:
    long_df["serie_key"] = long_df[key_cols].astype(str).agg(" | ".join, axis=1)
else:
    long_df["serie_key"] = "TOTAL"


agg_df = (long_df
          .groupby(["serie_key", "periodo_ref"], as_index=False)["valor"].sum()
          .sort_values(["serie_key", "periodo_ref"]))
def add_rolling(df):
    df = df.sort_values("periodo_ref")
    df["mm3"]  = df["valor"].rolling(3, min_periods=1).mean()
    df["mm6"]  = df["valor"].rolling(6, min_periods=1).mean()
    df["mm12"] = df["valor"].rolling(12, min_periods=1).mean()
    return df

agg_df = (agg_df
          .groupby("serie_key", group_keys=False)
          .apply(add_rolling)
          .reset_index(drop=True))

out_long = PROC_DIR / "series_long_t6_t61.csv"
agg_df.to_csv(out_long, index=False, encoding="utf-8")
print(f"💾 Séries (long) salvas em: {out_long}")

def plot_series(df, key=None, cols=("valor","mm3","mm6","mm12"), title_prefix="Série"):
    d = df if key is None else df[df["serie_key"] == key]
    if d.empty:
        print("Nada a plotar.")
        return
    x = pd.PeriodIndex(d["periodo_ref"], freq="M").to_timestamp()
    for c in cols:
        if c in d.columns:
            plt.figure(figsize=(9,4.5))
            plt.plot(x, d[c])
            plt.gca().yaxis.set_major_formatter(FuncFormatter(_fmt_mil))
            plt.title(f"{title_prefix} — {key if key else 'TOTAL (todas as chaves)'} — {c}")
            plt.xlabel("Período"); plt.ylabel("Postos")
            plt.tight_layout(); plt.show()

def plot_stl(df, key=None):
    d = df if key is None else df[df["serie_key"] == key]
    if d.empty:
        print("Nada para STL.")
        return
    d = d.copy().sort_values("periodo_ref")
    idx = pd.PeriodIndex(d["periodo_ref"], freq="M").to_timestamp()
    y = pd.Series(d["valor"].values, index=idx).asfreq("MS")
    y = y.interpolate()
    stl = STL(y, period=12, robust=True).fit()

    )
    for comp_name, comp in [("Observado", stl.observed), ("Tendência", stl.trend),
                            ("Sazonal", stl.seasonal), ("Resíduo", stl.resid)]:
        plt.figure(figsize=(9,4))
        plt.plot(comp.index, comp.values)
        plt.gca().yaxis.set_major_formatter(FuncFormatter(_fmt_mil))
        plt.title(f"STL — {comp_name} — {key if key else 'TOTAL'}")
        plt.xlabel("Período"); plt.ylabel("Valor")
        plt.tight_layout(); plt.show()

def plot_acf_pacf(df, key=None, lags=24):
    d = df if key is None else df[df["serie_key"] == key]
    if d.empty:
        print("Nada para ACF/PACF.")
        return
    d = d.copy().sort_values("periodo_ref")
    idx = pd.PeriodIndex(d["periodo_ref"], freq="M").to_timestamp()
    y = pd.Series(d["valor"].values, index=idx).asfreq("MS").fillna(0)

    plt.figure(figsize=(9,4))
    plot_acf(y, lags=min(lags, len(y)-1))
    plt.title(f"ACF — {key if key else 'TOTAL'}")
    plt.tight_layout(); plt.show()

    plt.figure(figsize=(9,4))
    plot_pacf(y, lags=min(lags, len(y)-1), method="ywm")
    plt.title(f"PACF — {key if key else 'TOTAL'}")
    plt.tight_layout(); plt.show()

sample_key = agg_df["serie_key"].iloc[0] if not agg_df.empty else None
print("Exemplo de chave para plot:", sample_key)

plot_series(agg_df, key=sample_key, cols=("valor","mm3","mm6","mm12"), title_prefix="Emprego Formal (Saldo)")
plot_stl(agg_df, key=sample_key)
plot_acf_pacf(agg_df, key=sample_key, lags=24)

print("\n🎯 Séries históricas geradas com sucesso: médias móveis, STL e ACF/PACF prontos.")
```

# **Modelo base**

<div align="justify"> <p> A etapa de modelagem deste projeto tem como propósito transformar a série temporal de vínculos empregatícios composta por admissões, desligamentos, saldo e estoque em informações analíticas que permitam identificar padrões, sazonalidades e tendências do mercado de trabalho formal brasileiro. Para isso, serão aplicadas técnicas estatísticas clássicas e interpretáveis, adequadas a séries mensais e a horizontes de previsão de curto prazo.

**4.1. Modelos de Referência (Benchmarks)**<p>
Como ponto de partida, serão aplicados os modelos os modelos Naive e Sazonal Naive:
*   Naive: considera o valor do período anterior como previsão para o próximo, sendo útil como linha de base para séries sem forte sazonalidade.
*   Sazonal Naive: replica o valor do mesmo mês do ano anterior, adequado para séries com padrões mensais recorrentes (como o aumento de contratações no comércio em dezembro).

Esses modelos simples têm função comparativa, permitindo avaliar a efetividade de abordagens mais sofisticadas.

**4.2. Decomposição Clássica** <p>
A decomposição clássica (STL – Seasonal-Trend decomposition using Loess) será utilizada para separar a série em três componentes fundamentais:
* Tendência: evolução de longo prazo do emprego formal no país;
* Sazonalidade: variações cíclicas recorrentes ao longo dos meses;
* Resíduo: flutuações aleatórias ou choques conjunturais.<p>

Esse procedimento, descrito por Cleveland et al. (1990), permitirá identificar padrões estruturais, como períodos de expansão e retração do emprego, e servirá como base para as etapas de suavização e previsão.

**4.3. Suavização Exponencial (Holt-Winters)** <p>
A técnica de Holt-Winters será empregada para capturar tendências e sazonalidades de forma adaptativa. Serão avaliadas as versões:
* Aditiva: quando a amplitude sazonal é constante;
*	Multiplicativa: quando a amplitude sazonal varia conforme o nível da série.
Esse método apresenta boa acurácia em horizontes curtos e interpretação intuitiva, sendo amplamente recomendado para séries econômicas mensais (HYNDMAN; ATHANASOPOULOS, 2018).

**4.4. Modelo ARIMA** <p>
Em séries que apresentarem forte autocorrelação temporal, será explorado o modelo ARIMA (AutoRegressive Integrated Moving Average), conforme Box, Jenkins e Reinsel (2015). Ele combina três componentes:<p>
* AR (AutoRegressivo): dependência entre observações passadas;
* I (Integrated): diferenciações para garantir estacionariedade;
* MA (Média Móvel): modelagem dos erros passados.<p>

O ARIMA será ajustado apenas após testes de estacionariedade (ADF) e de autocorrelação (ACF/PACF). Seu uso será complementar ao Holt-Winters, reforçando a robustez analítica do projeto.

**4.5. Métricas de Avaliação** <p>
Para avaliar o desempenho dos modelos, serão aplicadas métricas padronizadas na literatura de séries temporais:<p>
* MAE (Mean Absolute Error): média dos erros absolutos;
*	RMSE (Root Mean Square Error): penaliza maiores desvios;
*	MAPE (Mean Absolute Percentage Error): expressa o erro em termos percentuais.
Essas métricas serão utilizadas tanto para avaliar o ajuste quanto para comparar a performance entre os modelos (HYNDMAN; KOEHLER, 2006).

**4.6. Validação Institucional** <p>
Além da validação estatística, os resultados serão comparados com os valores oficiais divulgados no Sumário Executivo do Novo CAGED (BRASIL, 2025). Essa etapa assegura que as análises estejam coerentes com as fontes oficiais, reforçando a credibilidade e aplicabilidade prática do pipeline desenvolvido.

**4.7. Justificativa da Escolha Metodológica** <p>
As técnicas selecionadas equilibram simplicidade, interpretabilidade e aderência aos objetivos do projeto:<p>
* Permitem compreender o comportamento da série temporal (tendência e sazonalidade);
* Geram previsões de curto prazo para subsidiar gestores públicos e privados;
* Mantêm o pipeline reprodutível, sem dependência de modelos complexos ou opacos. <p>

**4.8.Avaliação do modelo base** <p>
Embora modelos híbridos (como ARIMA + LSTM ou GRU) apresentem ganhos de acurácia (ZHANG, 2003; LI et al., 2021), eles exigem maior volume de dados e infraestrutura computacional. Por esse motivo, o projeto opta por abordagens clássicas e consolidadas, mais compatíveis com o escopo atual.

A avaliação do modelo base no projeto foi conduzida com rigor estatístico e aderência à proposta de compreender o comportamento das séries temporais do Novo CAGED (2020–2025). O objetivo foi identificar o modelo mais adequado para descrever e projetar o saldo de empregos formais no Brasil, a partir das bases integradas no pipeline e validadas com o Sumário Executivo – Junho de 2025.

A escolha metodológica priorizou modelos estatísticos clássicos — em detrimento de modelos de aprendizado de máquina supervisionados — por se tratarem de séries econômicas curtas, mensais e altamente sazonais. Quatro métodos foram implementados e comparados:

* Naive: usa o último valor observado como previsão do próximo período, servindo como referência mínima de desempenho;

* Sazonal Naive: projeta o valor do mês corrente como igual ao do mesmo mês do ano anterior, capturando efeitos sazonais simples;

* Holt-Winters (Aditivo): aplica suavização exponencial, modelando tendência e sazonalidade de maneira adaptativa;

* SARIMA (2,1,0)(1,1,0)[12]: combina componentes autorregressivos, integração e médias móveis, ajustadas à periodicidade anual da série.

Os dados foram divididos em 60 observações para treino (2020–2024) e 6 para teste (janeiro a junho de 2025), garantindo avaliação robusta da capacidade preditiva.
As métricas MAE (Erro Absoluto Médio), RMSE (Raiz do Erro Quadrático Médio) e MAPE (Erro Percentual Absoluto Médio) foram calculadas para comparar o desempenho dos modelos.

Os resultados demonstraram melhor desempenho do modelo SARIMA, com MAE ≈ 84.775, RMSE ≈ 100.587 e MAPE ≈ 0,18%, superando os demais métodos. O modelo Holt-Winters apresentou boa estabilidade, mas leve superestimação em pontos de virada. Já os modelos Naive e Sazonal Naive foram mantidos como benchmarks comparativos, confirmando a relevância da modelagem paramétrica para séries com forte componente sazonal.

A análise dos resíduos revelou ausência de autocorrelação significativa, indicando adequação do modelo SARIMA. A decomposição da série mostrou tendência positiva sustentada desde 2022 e sazonalidade anual pronunciada em dezembro, associada ao setor de comércio.

Por fim, realizou-se a validação institucional comparando o saldo previsto para junho de 2025 com o valor oficial do Sumário Executivo do MTE. O resultado foi idêntico (166.621 vínculos), confirmando a consistência do modelo e a fidelidade do pipeline.

A escolha do SARIMA como modelo base justificou-se por sua capacidade de capturar tendência e sazonalidade de forma interpretável e estatisticamente estável, adequando-se ao escopo do projeto e às diretrizes metodológicas de transparência e reprodutibilidade científica.

```python
import warnings, itertools
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from matplotlib.ticker import FuncFormatter
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from IPython.display import display, HTML


SEASONAL_PERIODS = 12
H = 6
desired_key = None
show_residual_diag = True

PROC_DIR = Path("/content/processed")
series_path = PROC_DIR / "series_long_t6_t61.csv"
assert series_path.exists(), "Não encontrei /content/processed/series_long_t6_t61.csv. Rode a Etapa 3 (parte 2)."

df = pd.read_csv(series_path)
assert {"serie_key","periodo_ref","valor"}.issubset(df.columns), "Arquivo de séries inválido (faltam colunas: 'serie_key','periodo_ref','valor')."
keys = df["serie_key"].astype(str).unique().tolist()

def is_note_key(k: str) -> bool:
    kx = str(k).strip().lower()
    if not kx or kx.startswith("*"):
        return True
    banned = [
        "variação mensal","variacao mensal","resultados acrescidos",
        "estoque total considera","estoque da competência","estoque da competencia",
        "referência do ano","referencia do ano","nota","observação","observacoes",
        "sem ajustes","com ajustes","fonte:","não identificado","nao identificado"
    ]
    if any(b in kx for b in banned):
        return True
    return len(kx) > 80

clean_keys = [k for k in keys if not is_note_key(k)]
print("Chaves (limpas) - exemplo das 20 primeiras:", clean_keys[:20])

total_key = next((k for k in keys if str(k).strip().lower() == "total"), None)
if desired_key and desired_key in keys:
    key = desired_key
elif total_key is not None:
    key = total_key
elif clean_keys:
    key = clean_keys[0]
else:
    key = keys[0]

print("Usando serie_key:", key)

s = (df[df["serie_key"] == key]
     .loc[:, ["periodo_ref","valor"]]
     .copy())

s["periodo_ref"] = pd.PeriodIndex(s["periodo_ref"], freq="M").astype(str)
s = (s.groupby("periodo_ref", as_index=False)["valor"]
       .sum()
       .sort_values("periodo_ref"))
idx = pd.PeriodIndex(s["periodo_ref"], freq="M").to_timestamp()
y = pd.Series(s["valor"].values, index=idx).sort_index()

y = y.asfreq("MS")
y = y.interpolate()

print("Período da série:", y.index.min().date(), "→", y.index.max().date(), "| pontos:", len(y))

min_needed = SEASONAL_PERIODS + 3
if len(y) <= min_needed:
    H = max(3, min(6, len(y) // 4))
    print(f" Série curta. Ajustando H automaticamente para {H}.")
assert len(y) > H + 3, f"Série muito curta ({len(y)} pts). Reduza H ou gere mais histórico."

# split treino/teste
y_train = y.iloc[:-H]
y_test  = y.iloc[-H:]
print(f"Tamanho treino: {len(y_train)}, teste: {len(y_test)} (H={H})")

def _fmt_mil(x, pos):
    try:
        return f"{int(x):,}".replace(",", ".")
    except:
        return str(x)


# Métricas
def mae(y_true, y_pred):
    return float(np.mean(np.abs(y_true - y_pred)))

def rmse(y_true, y_pred):
    return float(np.sqrt(np.mean((y_true - y_pred)**2)))

def mape(y_true, y_pred):
    y_true = np.asarray(y_true, dtype=float)
    y_pred = np.asarray(y_pred, dtype=float)
    denom = np.where(y_true == 0, np.nan, np.abs(y_true))
    return float(np.nanmean(np.abs((y_true - y_pred)/denom)) * 100.0)

def eval_metrics(name, y_true, y_pred):
    return {"modelo": name, "MAE": mae(y_true, y_pred), "RMSE": rmse(y_true, y_pred), "MAPE_%": mape(y_true, y_pred)}

results = []
fcst_df = pd.DataFrame(index=y_test.index)


# 1) Naive

last_val = y_train.iloc[-1]
yhat_naive = pd.Series([last_val]*H, index=y_test.index)
fcst_df["Naive"] = yhat_naive
results.append(eval_metrics("Naive", y_test, yhat_naive))


# 2) Sazonal Naive (fallback se histórico < 12)
if len(y_train) >= SEASONAL_PERIODS:
    base = y_train.iloc[-SEASONAL_PERIODS:]
    reps = int(np.ceil(H / SEASONAL_PERIODS))
    vals = np.tile(base.values, reps)[:H]
    yhat_snaive = pd.Series(vals, index=y_test.index)
    fcst_df["SazonalNaive"] = yhat_snaive
    results.append(eval_metrics("Sazonal Naive", y_test, yhat_snaive))
else:
    print(" Sazonal Naive indisponível (histórico < 12). Mantido somente Naive.")


# 3) Holt-Winters (heurística para 'add' vs 'mul')
def seasonal_mode(series, m=SEASONAL_PERIODS):
    # Heurística: se amplitude sazonal cresce com o nível, usar multiplicativo
    if len(series) < 2*m:
        return "add"
    try:
        trend = series.rolling(m, min_periods=1).mean()
        seas = (series - trend).abs().rolling(m, min_periods=1).mean()
        ratio = (seas / (trend.abs() + 1e-6)).median()
        return "mul" if ratio > 0.2 else "add"
    except Exception:
        return "add"

seasonal_type = seasonal_mode(y_train, SEASONAL_PERIODS)
try:
    hw = ExponentialSmoothing(
        y_train,
        trend="add",
        seasonal=seasonal_type,
        seasonal_periods=SEASONAL_PERIODS,
        initialization_method="estimated"
    ).fit(optimized=True, use_brute=True)
    yhat_hw = hw.forecast(H)
    fcst_df["HoltWinters"] = yhat_hw
    results.append(eval_metrics(f"Holt-Winters ({seasonal_type})", y_test, yhat_hw))
except Exception as e:
    print(" Holt-Winters falhou:", e)


# 4) SARIMA (busca enxuta por AIC, com limite de complexidade)
def fit_best_sarima(y_train, seasonal_periods=12):
    p = d = q = [0,1,2]
    P = D = Q = [0,1]
    seasonal_orders = [(P_, D_, Q_, seasonal_periods) for P_ in P for D_ in D for Q_ in Q]
    orders = list(itertools.product(p, d, q))

    best = {"aic": np.inf, "order": None, "seasonal_order": None, "model": None}
    for order in orders:
        for seas in seasonal_orders:
            if sum(order) + sum(seas[:3]) > 5:
                continue
            try:
                mod = SARIMAX(
                    y_train,
                    order=order,
                    seasonal_order=seas,
                    enforce_stationarity=False,
                    enforce_invertibility=False
                ).fit(disp=False)
                if np.isfinite(mod.aic) and mod.aic < best["aic"]:
                    best = {"aic": mod.aic, "order": order, "seasonal_order": seas, "model": mod}
            except Exception:
                continue
    return best

best = fit_best_sarima(y_train, SEASONAL_PERIODS)
if best["model"] is not None:
    sarima = best["model"]
    yhat_sarima = sarima.forecast(H)
    fcst_df["SARIMA"] = yhat_sarima
    results.append(eval_metrics(f"SARIMA {best['order']} x {best['seasonal_order']}", y_test, yhat_sarima))
    print("✔️ Melhor SARIMA:", best["order"], "x", best["seasonal_order"], "| AIC:", round(best["aic"], 2))
else:
    print("ℹ️ SARIMA não encontrado em busca enxuta; pulando.")

# Tabela de resultados (formatada — MAPE em %)
res_df = pd.DataFrame(results).sort_values("RMSE")

res_df_fmt = res_df.copy()
res_df_fmt["MAE"]  = res_df_fmt["MAE"].round(0).astype(int)
res_df_fmt["RMSE"] = res_df_fmt["RMSE"].round(0).astype(int)
# Se MAPE veio como fração (<1), converte para %
if (res_df_fmt["MAPE_%"] < 1).all():
    res_df_fmt["MAPE_%"] = (res_df_fmt["MAPE_%"] * 100).round(2)
else:
    res_df_fmt["MAPE_%"] = res_df_fmt["MAPE_%"].round(2)

display(HTML("<h3>📊 Desempenho dos Modelos (teste)</h3>"))
display(res_df_fmt)


# Gráfico comparativo (treino/teste + previsões)
plt.figure(figsize=(11,5))
plt.plot(y_train.index, y_train.values, label="Treino")
plt.plot(y_test.index,  y_test.values,  label="Teste")
for col in fcst_df.columns:
    plt.plot(fcst_df.index, fcst_df[col].values, label=col)
plt.gca().yaxis.set_major_formatter(FuncFormatter(_fmt_mil))
plt.title(f"Comparativo de Previsões — {key}")
plt.xlabel("Período"); plt.ylabel("Postos (saldo)")
plt.legend()
plt.tight_layout(); plt.show()


# Diagnóstico de resíduos (modelo campeão por RMSE) — lags seguros
best_name = res_df.iloc[0]["modelo"]
print("Melhor modelo (RMSE):", best_name)

def map_model_to_col(model_name: str) -> str | None:
    m = model_name.lower()
    if "naive" in m and "sazonal" not in m:
        return "Naive"
    if "sazonal" in m:
        return "SazonalNaive"
    if "holt" in m:
        return "HoltWinters"
    if "sarima" in m:
        return "SARIMA"
    return None

chosen_col = map_model_to_col(best_name)
if chosen_col in fcst_df.columns:
    resid = (y_test - fcst_df[chosen_col]).dropna()
else:
    resid = (y_test - fcst_df.iloc[:, 0]).dropna()

if show_residual_diag and len(resid) >= 3:
    plt.figure(figsize=(9,4))
    plt.plot(resid.index, resid.values)
    plt.gca().yaxis.set_major_formatter(FuncFormatter(_fmt_mil))
    plt.title(f"Resíduos — {best_name}")
    plt.xlabel("Período"); plt.ylabel("Erro")
    plt.tight_layout(); plt.show()

    # lags seguros: <= 50% do tamanho da amostra e no mínimo 1
    safe_lags = max(1, min(12, (len(resid) // 2) - 1))
    if safe_lags >= 1:
        plt.figure(figsize=(9,4))
        plot_acf(resid, lags=safe_lags)
        plt.title("ACF dos Resíduos")
        plt.tight_layout(); plt.show()

        plt.figure(figsize=(9,4))
        try:
            plot_pacf(resid, lags=safe_lags, method="ywm")
            plt.title("PACF dos Resíduos")
            plt.tight_layout(); plt.show()
        except Exception as e:
            print(" PACF não gerado (amostra curta):", e)
    else:
        print(" Amostra de resíduos muito pequena para ACF/PACF — pulando.")
else:
    print(" Amostra de resíduos insuficiente para diagnóstico — pulando.")
```
